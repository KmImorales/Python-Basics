## Ordering the Bars for Analysis

8. The graphs are readable, but you can take it another step further.
   - Add the `order` parameter to your `.countplot()` method.
   - Set the parameter to that the value counts in each column are in descending order.
   
   You will need to use the `.value_counts()` pandas method and the `.index` pandas object.

````python

# Define the number of rows and columns for the subplot grid
num_columns = len(df.columns)
cols = 5  # Adjust number of columns in the grid
rows = (num_columns + cols - 1) // cols  # Calculate number of rows needed

# Create a figure and axes with the specified grid layout
fig, axes = plt.subplots(rows, cols, figsize=(20, 4 * rows))  # Adjust overall figure size

# Flatten axes array to iterate over it more easily
axes = axes.flatten()

# Iterate through the columns and create countplots
for i, column in enumerate(df.columns):
    # Get the value counts for the column and order them in descending order
    order = df[column].value_counts().index
    
    # Create the countplot with the specified order
    sns.countplot(x=column, data=df, ax=axes[i], order=order)
    
    # Set the title and rotate x-axis labels for better readability
    axes[i].set_title(column)
    axes[i].tick_params(axis='x', rotation=45)

# Turn off any unused axes (if any)
for i in range(num_columns, len(axes)):
    axes[i].axis('off')

# Show the plot
plt.tight_layout()  # Adjust the layout to prevent overlap
plt.show()
````

## Extensions

10. Feel free to play around with the graphs and customize them any way you want to help in your analysis! Here are some ideas to get yourself started:
    * Turn any bar graph with less than six bars into a pie chart (hint: use a conditional statement!).
    * Create your bar charts using a list comprehension instead of a `for` loop.
    * Change the color theme of your graphs using the seaborn [color](https://seaborn.pydata.org/generated/seaborn.countplot.html) or [palette](https://seaborn.pydata.org/tutorial/color_palettes.html) parameters.
    * Remove any graphs you find uninformative.
    * Change around the current title or label formatting.
    
    Happy Coding! :)

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# load in the data
df = pd.read_csv("mushroom_data.csv")

# convert 'Bruises' column to strings if necessary
df['Bruises'] = df['Bruises'].apply(str)

# Define the number of rows and columns for the subplot grid
num_columns = len(df.columns)
cols = 5  # Adjust number of columns in the grid
rows = (num_columns + cols - 1) // cols  # Calculate number of rows needed

# Create a figure and axes with the specified grid layout
fig, axes = plt.subplots(rows, cols, figsize=(20, 4 * rows))  # Adjust overall figure size
axes = axes.flatten()

# Use list comprehension to create countplots (with conditional check for pie chart)
for i, column in enumerate(df.columns):
    # Skip columns with no variation or too many categories
    if df[column].nunique() > 10:
        continue  # Skip columns with more than 10 unique values, adjust as needed
    
    # Check if there are fewer than 6 unique values (change to pie chart if true)
    if df[column].nunique() < 6:
        # Plot a pie chart instead of a bar chart
        df[column].value_counts().plot(kind='pie', ax=axes[i], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
        axes[i].set_title(f"{column} (Pie Chart)")
    else:
        # Create a countplot with custom color palette using 'hue' instead of 'palette'
        order = df[column].value_counts().index
        sns.countplot(x=column, data=df, ax=axes[i], order=order, hue=column, palette="Set2", legend=False)
        axes[i].set_title(f"{column} (Bar Chart)")
    
    # Rotate x-axis labels for readability
    axes[i].tick_params(axis='x', rotation=45)

    # Label formatting
    axes[i].set_xlabel(f"{column} values")
    axes[i].set_ylabel("Count")

# Turn off any unused axes (if any)
for i in range(num_columns, len(axes)):
    axes[i].axis('off')

# Show the plot
plt.tight_layout()  # Adjust the layout to prevent overlap
plt.show()
````
